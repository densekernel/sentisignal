> For the dataset grouped by stock exchange: do sentiment summary based on messages quantity

    1. select columns: bull messages, bear messages and total scanned messages 
    2. calculate the quantity of neutral messages
    4. save the messages table, named "messagesTable"
    5. transform the quantity numbers to percentage numbers and format by %
    6. store the summary on "sentiOutcome" and export to 'sentimentSummaryByExchange.csv'

```python
# a table only contains the total quality of bullish, bearish and total messages for each exchange
messagesTable = sumByExchange.ix[:,[3,4,6]]
messagesTable.head()
```

```python
# neutral = total - bull - bear
neutralCounts = messagesTable['TOTAL_SCANNED_MESSAGES']-messagesTable['BULL_SCORED_MESSAGES']-messagesTable['BEAR_SCORED_MESSAGES']
# print (neutralCounts)
```

```python
# add new columns to the table. by using 'insert(idx, col_name, value)'
messagesTable.insert(2, 'NEUTRAL_SCORED_MESSAGES', neutralCounts)
messagesTable.head()
```

```python
# the percentage of bull messages 
bullPerc = messagesTable['BULL_SCORED_MESSAGES']/messagesTable['TOTAL_SCANNED_MESSAGES']
# the percentage of bear messages
bearPerc = messagesTable['BEAR_SCORED_MESSAGES']/messagesTable['TOTAL_SCANNED_MESSAGES']
# compute percentage of neutral messages quantity
neutralPerc = neutralCounts/messagesTable['TOTAL_SCANNED_MESSAGES']
```

```python
# change data format to %
bullPerc = bullPerc.apply(lambda x: format(x, '.2%'))
bearPerc = bearPerc.apply(lambda x: format(x, '.2%'))
neutralPerc = neutralPerc.apply(lambda x: format(x, '.2%'))
```

```python
# sentiment summary by stock exchange
sentiOutcome = messagesTable
sentiOutcome.insert(1,'BULL_PERC', bullPerc)
sentiOutcome.insert(3,'BEAR_PERC', bearPerc)
sentiOutcome.insert(5,'NEUTRAL_PERC', neutralPerc)
sentiOutcome.head()
```

```python
# export to a csv file
sentiOutcome.to_csv('~/psychSignalData/output/sentimentSummaryByExchange.csv', encoding='utf-8')
```

> Dataset merge on business sectors.

    a. merge dailyData set to every sector
    b. remove empty rows for each merged sector dataset
    c. merge the sectors datasets together to 'sectorsData' and export to 'DataBySectors.csv'
    d. count the available rows for each sector
    
        1. material:   707,298;

    e. statistic summary based on sectors, exchanges and companies, stored in 'stat_groupedSeExCom'
    f. then output is exported to 'StatisticSectorsExCom_daily.csv'

```python
# merge
# merge dataset right join to sector business
secMatMerge = merge(dailyData,secMat,on='SYMBOL',how = "right")
# secMatMerge.isnull()
secMatMerge = secMatMerge.dropna()
secMatMerge['BUSINESS_SECTOR'] = 'Material'
secMatMerge.info()
```

```python
secCongMerge = merge(dailyData,secCong,on='SYMBOL',how = "right")
secCongMerge = secCongMerge.dropna()
secCongMerge['BUSINESS_SECTOR'] = 'Conglomerates'
secCongMerge.info()
```

```python
secConsuMerge = merge(dailyData,secConsu,on='SYMBOL',how = "right")
secConsuMerge = secConsuMerge.dropna()
secConsuMerge['BUSINESS_SECTOR'] = 'Consumer_goods'
secConsuMerge.info()
```

```python
secFinanMerge = merge(dailyData,secFinan,on='SYMBOL',how = "right")
secFinanMerge = secFinanMerge.dropna()
secFinanMerge['BUSINESS_SECTOR'] = 'Financial'
secFinanMerge.info()
```

```python
secHealthMerge = merge(dailyData,secHealth,on='SYMBOL',how = "right")
secHealthMerge = secHealthMerge.dropna()
secHealthMerge['BUSINESS_SECTOR'] = 'Healthcare'
secHealthMerge.info()
```

```python
secIndusMerge = merge(dailyData,secIndus,on='SYMBOL',how = "right")
secIndusMerge = secIndusMerge.dropna()
secIndusMerge['BUSINESS_SECTOR'] ='Industrial_goods'
secIndusMerge.info()
```

```python
secServMerge = merge(dailyData,secServ,on='SYMBOL',how = "right")
secServMerge = secServMerge.dropna()
secServMerge['BUSINESS_SECTOR'] = 'Services'
secServMerge.info()
```

```python
secTechMerge = merge(dailyData,secTech,on='SYMBOL',how = "right")
secTechMerge = secTechMerge.dropna()
secTechMerge['BUSINESS_SECTOR'] ='Technology'
secTechMerge.info()
```

```python
secUtilMerge = merge(dailyData,secUtil,on='SYMBOL',how = "right")
secUtilMerge = secUtilMerge.dropna()
secUtilMerge['BUSINESS_SECTOR'] = 'Utilities'
secUtilMerge.info()
```

```python
# merge the sectors datasets together
sectorsData = concat([secMatMerge,secCongMerge,secConsuMerge,secFinanMerge,secHealthMerge,
                      secIndusMerge,secServMerge,secTechMerge,secUtilMerge],
                     keys=['Matrial','Conglomerates','Consumer_goods','Financial','Healthcare',
                           'Industrial_goods','Services','Technology','Utilities'],
                     names = ['SECTOR','Sector_Index'])
sectorsData.head(100)
```

```python
sectorsData.to_csv('~/psychSignalData/output/DataBySectors.csv', encoding='utf-8')
```
